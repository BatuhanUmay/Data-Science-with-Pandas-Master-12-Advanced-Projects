## The best and the worst movies...

2. __Filter__ the Dataset and __find the best/worst n Movies__ with the

- Highest Revenue
- Highest Budget
- Highest Profit (=Revenue - Budget)
- Lowest Profit (=Revenue - Budget)
- Highest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 
- Lowest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10)
- Highest number of Votes
- Highest Rating (only movies with 10 or more Ratings)
- Lowest Rating (only movies with 10 or more Ratings)
- Highest Popularity

__Define__ an appropriate __user-defined function__ to reuse code.

```python
df["profit_musd"] = df["revenue_musd"] - df["budget_musd"]
# df["return"] = df["revenue_musd"] / df["budget_musd"]
df["ROI"] = df["revenue_musd"] / df["budget_musd"]
```

```python
from IPython.display import HTML

mydf = df[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]].copy()
mydf["profit_musd"] = df.revenue_musd.sub(df.budget_musd)
mydf["return"] = df.revenue_musd.div(df.budget_musd)
mydf.columns = ["Url", "Title", "Budget", "Revenue", "Votes",
                   "Average Rating", "Popularity", "Profit", "ROI"]
mydf.set_index("Title", inplace = True)
mydf.Budget.fillna(0, inplace = True)
mydf.Votes.fillna(0, inplace = True)
mydf
```

```python
def best_worst(n, by, min_bud=0, min_votes=0, ascending=False):
    df2 = mydf.loc[(mydf.Budget >= min_bud) & (mydf.Votes >= min_votes), ["Url", by]].sort_values(by=by, ascending=ascending).head(n).copy()
    return HTML(df2.to_html(escape=False))
```

__Movies Top 5 - Highest Revenue__

```python
best_worst(n = 5, by = "Revenue")
```

__Movies Top 5 - Highest Budget__

```python
best_worst(5, "Budget")
```

__Movies Top 5 - Highest Profit__

```python
best_worst(5, "Profit")
```

__Movies Top 5 - Lowest Profit__

```python
best_worst(5, "Profit", ascending = True)
```

__Movies Top 5 - Highest ROI__

```python
best_worst(5, "ROI", min_bud = 50)
```

__Movies Top 5 - Lowest ROI__

```python
best_worst(5, "ROI", ascending = True, min_bud = 100)
```

```python
best_worst(5, "Average Rating", ascending = True, min_votes = 100)
```

## Find your next Movie

3. __Filter__ the Dataset for movies that meet the following conditions:

__Search 1: Science Fiction Action Movie with Bruce Willis (sorted from high to low Rating)__

```python
mask_genres = df.genres.str.contains("Science Fiction") & df.genres.str.contains("Action")
mask_actor = df.cast.str.contains("Bruce Willis")

bruce = df.loc[mask_genres & mask_actor, ["title", "poster_path", "vote_average"]].sort_values(by="vote_average", ascending=False).set_index("title")
HTML(bruce.to_html(escape=False))
```

__Search 2: Movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)__

```python
mask_actor = df.cast.str.contains("Uma Thurman")
mask_director = df.director.str.contains("Quentin Tarantino")
tarantino = df.loc[mask_actor & mask_director, ["title", "poster_path", "runtime"]].sort_values(by="runtime", ascending=False).set_index("title")
HTML(tarantino.to_html(escape=False))
```

__Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)__

```python
mask_production_companies = df.production_companies.str.contains("Pixar").fillna(False)
# mask_year = df.release_date.between("2010-01-01", "2015-12-31")
mask_year = (df.release_date.dt.year <= 2015) & (df.release_date.dt.year >= 2010)
pixar = df.loc[mask_production_companies & mask_year, ["title", "poster_path", "revenue_musd"]].sort_values(by="revenue_musd", ascending=False).set_index("title")
HTML(pixar.to_html(escape=False))
```

__Search 4: Action or Thriller Movie with original language English and minimum Rating of 7.5 (most recent movies first)__

```python
df.head(1)
```

```python
mask_genres = df.genres.str.contains("Action") | df.genres.str.contains("Thriller")
mask_original_language = df.original_language == "en"
mask_rating = df.vote_average >= 7.5
mask_vote_count = df.vote_count >= 10
filtered = df.loc[mask_genres & mask_original_language & mask_rating & mask_vote_count, ["title", "poster_path", "genres", "vote_average", "vote_count", "release_date"]].sort_values(by = "release_date", ascending = False).set_index("title").head(20)
HTML(filtered.to_html(escape=False))
```

## What are the most common Words in Movie Titles and Taglines?

```python
taglines = {}
titles = {}

for i in df["tagline"]:
    for word in str(i).split():
        if word != "nan":
            if word in taglines:
                taglines[word] += 1
            else:
                taglines[word] = 1

for i in df["title"]:
    for word in str(i).split():
        if word != "nan":
            if word in titles:
                titles[word] += 1
            else:
                titles[word] = 1

print(sorted(taglines.items(), key=lambda x: x[1], reverse=True)[:10])
print(sorted(titles.items(), key=lambda x: x[1], reverse=True)[:10])
```
